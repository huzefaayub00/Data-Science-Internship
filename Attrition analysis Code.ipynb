```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Data Collection
data = pd.read_csv('Attrition data.csv')

# Step 2: Data Cleaning and Preparation
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)
data= data.replace({'Attrition': {'Yes': True,'No': False}}) 

# Step 3: Exploratory Data Analysis (EDA)
print(data.describe())

plt.figure(figsize=(10, 6))
sns.lineplot(x='YearsAtCompany', y='Attrition', data=data)
plt.title('Attrition Rate vs. Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Attrition')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Step 4: Identifying Key Factors
#data= data.replace({'Attrition': {'Yes': True,'No': False}}) 
correlation_matrix = data.corr()
attrition_correlation = correlation_matrix['Attrition'].sort_values(ascending=False)
print(attrition_correlation)

# Step 5: Dashboard Design and Development
def dashboard_overview():
    plt.figure(figsize=(10, 6))
    sns.barplot(x=attrition_correlation.index, y=attrition_correlation.values)
    plt.title('Correlation of Features with Attrition (Overview)')
    plt.xlabel('Feature')
    plt.ylabel('Correlation with Attrition')
    plt.xticks(rotation=45)
    plt.show()

def dashboard_breakdown():
    # Select top N features with highest positive correlation and top N features with lowest negative correlation
    N = 5
    top_positive_features = attrition_correlation.head(N)
    top_negative_features = attrition_correlation.tail(N)
    
    # Concatenate top positive and top negative features
    top_features = pd.concat([top_positive_features, top_negative_features])

    plt.figure(figsize=(12, 8))
    sns.barplot(x=top_features.index, y=top_features.values)
    plt.title('Top Features with Highest Positive and Negative Correlation with Attrition (Breakdown)')
    plt.xlabel('Feature')
    plt.ylabel('Correlation with Attrition')
    plt.xticks(rotation=45)
    plt.show()

dashboard_overview()
dashboard_breakdown()
```
